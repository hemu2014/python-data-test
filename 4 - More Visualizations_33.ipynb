#### The `subplot2grid` command

There is another way to make subplots using a grid-like format:

## rowspan与colspan指定图形占据的大小，第一个参数就是总的大小，第二个参数就是指定位置从哪开始绘制，索引默认从0开始

```python
plt.figure(figsize=(14, 6))

ax1 = plt.subplot2grid((3,3), (0,0), colspan=3)
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2)
ax3 = plt.subplot2grid((3,3), (1,2), rowspan=2)
ax4 = plt.subplot2grid((3,3), (2,0))
ax5 = plt.subplot2grid((3,3), (2,1))
```

### Scatter Plot

何时需要关心 s 的具体值？
跨图形一致性：如果需要在多个图中保持相同的大小比例，需固定 s 的计算方式。

绝对尺寸控制：若需精确控制点的物理尺寸（如导出为 PDF 时），需结合 fig.dpi 和 s 的换算。

6. 实用建议
直接使用半径的平方：s = r**2（更直观，避免冗余计算）。

调整系数控制大小范围：例如 s = 100 * np.random.rand(N)**2。

避免极端大小：过大的 s 会导致点重叠严重，过小则难以看清。

总结
π 不是必须的：s 参数本质是相对大小，比例正确即可。

原代码使用 np.pi 是为了符合数学公式的直观性，但实际绘图时可省略。

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi *(20 * np.random.rand(N))**2  # 0 to 15 point radii
```

```python
plt.figure(figsize=(14, 6))

plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Spectral')
plt.colorbar()

plt.show()
```

```python
fig = plt.figure(figsize=(14, 6))

ax1 = fig.add_subplot(1,2,1)
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel1')
plt.colorbar()

ax2 = fig.add_subplot(1,2,2)
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel2')
plt.colorbar()

plt.show()
```

```python
'''
==========================
Lightness of the colormaps
==========================

For each colormap, plot the lightness parameter L* from CIELAB colorspace
along the y axis vs index through the colormap. Colormaps are examined in
categories as in the original matplotlib gallery of colormaps.
'''

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import cm
from colorspacious import cspace_converter
from colormaps import cm  # the colormaps, grouped by category

mpl.rcParams.update({'font.size': 12})

# Number of colormap per subplot for particular cmap categories
_DSUBS = {'Perceptually Uniform Sequential': 4, 'Sequential': 6,
          'Sequential (2)': 6, 'Diverging': 6, 'Qualitative': 4,
          'Miscellaneous': 6}

# Spacing between the colormaps of a subplot
_DC = {'Perceptually Uniform Sequential': 1.4, 'Sequential': 0.7,
       'Sequential (2)': 1.4, 'Diverging': 1.4, 'Qualitative': 1.4,
       'Miscellaneous': 1.4}

# Indices to step through colormap
x = np.linspace(0.0, 1.0, 100)

# Do plot
for cmap_category, cmap_list in cm:

    # Do subplots so that colormaps have enough space.
    # Default is 6 colormaps per subplot.
    dsub = _DSUBS.get(cmap_category, 6)
    nsubplots = int(np.ceil(len(cmap_list) / float(dsub)))

    # squeeze=False to handle similarly the case of a single subplot
    fig, axes = plt.subplots(nrows=nsubplots, squeeze=False,
                             figsize=(7, 2.6*nsubplots))

    for i, ax in enumerate(axes.flat):

        locs = []  # locations for text labels

        for j, cmap in enumerate(cmap_list[i*dsub:(i+1)*dsub]):

            # Get RGB values for colormap and convert the colormap in
            # CAM02-UCS colorspace.  lab[0, :, 0] is the lightness.
            rgb = cm.get_cmap(cmap)(x)[np.newaxis, :, :3]
            lab = cspace_converter("sRGB1", "CAM02-UCS")(rgb)

            # Plot colormap L values.  Do separately for each category
            # so each plot can be pretty.  To make scatter markers change
            # color along plot:
            # http://stackoverflow.com/questions/8202605/matplotlib-scatterplot-colour-as-a-function-of-a-third-variable

            if cmap_category == 'Sequential':
                # These colormaps all start at high lightness but we want them
                # reversed to look nice in the plot, so reverse the order.
                y_ = lab[0, ::-1, 0]
                c_ = x[::-1]
            else:
                y_ = lab[0, :, 0]
                c_ = x

            dc = _DC.get(cmap_category, 1.4)  # cmaps horizontal spacing
            ax.scatter(x + j*dc, y_, c=c_, cmap=cmap, s=300, linewidths=0.0)

            # Store locations for colormap labels
            if cmap_category in ('Perceptually Uniform Sequential',
                                 'Sequential'):
                locs.append(x[-1] + j*dc)
            elif cmap_category in ('Diverging', 'Qualitative',
                                   'Miscellaneous', 'Sequential (2)'):
                locs.append(x[int(x.size/2.)] + j*dc)

        # Set up the axis limits:
        #   * the 1st subplot is used as a reference for the x-axis limits
        #   * lightness values goes from 0 to 100 (y-axis limits)
        ax.set_xlim(axes[0, 0].get_xlim())
        ax.set_ylim(0.0, 100.0)

        # Set up labels for colormaps
        ax.xaxis.set_ticks_position('top')
        ticker = mpl.ticker.FixedLocator(locs)
        ax.xaxis.set_major_locator(ticker)
        formatter = mpl.ticker.FixedFormatter(cmap_list[i*dsub:(i+1)*dsub])
        ax.xaxis.set_major_formatter(formatter)
        for label in ax.get_xticklabels():
            label.set_rotation(50)

    ax.set_xlabel(cmap_category + ' colormaps', fontsize=14)
    fig.text(0.0, 0.55, 'Lightness $L^*$', fontsize=12,
             transform=fig.transFigure, rotation=90)

    fig.tight_layout(h_pad=0.0, pad=1.5)
    plt.show()
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import cm
from colorspacious import cspace_converter

mpl.rcParams.update({'font.size': 12})

# 使用Matplotlib内置色图分类（替代原colormaps模块）
cmap_categories = {
    'Perceptually Uniform Sequential': ['viridis', 'plasma', 'inferno', 'magma', 'cividis'],
    'Sequential': ['Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds'],
    'Diverging': ['PiYG', 'PRGn', 'BrBG', 'PuOr', 'RdGy', 'RdBu'],
    'Qualitative': ['Pastel1', 'Pastel2', 'Paired', 'Accent'],
    'Miscellaneous': ['flag', 'prism', 'ocean', 'gist_earth', 'terrain', 'tab20c']
}

# 调整子图布局参数
_DSUBS = {'Perceptually Uniform Sequential': 4, 'Sequential': 6,
          'Diverging': 6, 'Qualitative': 4, 'Miscellaneous': 6}
_DC = {'Perceptually Uniform Sequential': 1.4, 'Sequential': 0.7,
       'Diverging': 1.4, 'Qualitative': 1.4, 'Miscellaneous': 1.4}

x = np.linspace(0.0, 1.0, 100)  # 色图索引

for cmap_category, cmap_list in cmap_categories.items():
    dsub = _DSUBS.get(cmap_category, 6)
    nsubplots = int(np.ceil(len(cmap_list) / dsub))

    fig, axes = plt.subplots(nrows=nsubplots, squeeze=False, figsize=(7, 2.6*nsubplots))

    for i, ax in enumerate(axes.flat):
        locs = []
        for j, cmap_name in enumerate(cmap_list[i*dsub:(i+1)*dsub]):
            cmap = cm.get_cmap(cmap_name)
            rgb = cmap(x)[np.newaxis, :, :3]  # 获取RGB值
            lab = cspace_converter("sRGB1", "CAM02-UCS")(rgb)  # 转换到CAM02-UCS

            if cmap_category == 'Sequential':
                y_ = lab[0, ::-1, 0]  # 反转亮度
                c_ = x[::-1]
            else:
                y_ = lab[0, :, 0]
                c_ = x

            dc = _DC.get(cmap_category, 1.4)
            ax.scatter(x + j*dc, y_, c=c_, cmap=cmap, s=300, linewidths=0.0)

            # 标记色图名称位置
            if cmap_category in ('Perceptually Uniform Sequential', 'Sequential'):
                locs.append(x[-1] + j*dc)
            else:
                locs.append(x[int(x.size/2.)] + j*dc)

        ax.set_xlim(axes[0, 0].get_xlim())
        ax.set_ylim(0.0, 100.0)
        ax.xaxis.set_ticks_position('top')
        ax.xaxis.set_major_locator(mpl.ticker.FixedLocator(locs))
        ax.xaxis.set_major_formatter(mpl.ticker.FixedFormatter(cmap_list[i*dsub:(i+1)*dsub]))
        for label in ax.get_xticklabels():
            label.set_rotation(50)

    ax.set_xlabel(cmap_category + ' colormaps', fontsize=14)
    fig.text(0.0, 0.55, 'Lightness $L^*$', fontsize=12, rotation=90)
    fig.tight_layout(h_pad=0.0, pad=1.5)

plt.show()
```

##使用zip()方法将两个数组组合在一起

为什么保留 i 即使未使用？
习惯写法：
enumerate 是 Python 中遍历“索引+元素”的标准方式，即使索引未使用也会保留。

未来扩展性：
方便后续需要按索引操作时直接调用，避免重构代码。

```python
fig, axes = plt.subplots(3, 2, figsize=(14, 12))  # 3行2列
axes = axes.flatten()  # 将二维轴数组展平为一维

bandwidths = [0.05, 0.1, 0.3, 'scott', 'silverman']

for i, (ax, bw) in enumerate(zip(axes[:5], bandwidths)):
    density.set_bandwidth(bw_method=bw)
    ax.plot(values2, density(values2), color='#FF7F00')
    ax.fill_between(values2, 0, density(values2), alpha=0.5, color='#FF7F00')
    ax.set_xlim(-5, 5)
    ax.set_title(f'Bandwidth = {bw}')
    ax.grid(True, linestyle='--', alpha=0.3)

# 隐藏多余的子图
for ax in axes[5:]:
    ax.axis('off')

plt.tight_layout()
plt.show()
```

### Combine plots

  ### counts: The frequency of data points in each bin.
每个分箱中数据点的频率。
  ### bins: The edges of the bins.
bins : 箱子的边缘。
  ### _: A third value that is ignored in this case (often patches representing the bars of the histogram).
_ : 在此情况下忽略的第三个值（通常是表示直方图条的补丁）。
● np.diff(bins) calculates the width of each bin.
计算每个箱子的宽度。
● counts * np.diff(bins) calculates the area of each bar (height * width).
counts * np.diff(bins) 计算每个条形的面积（高度 * 宽度）。
● np.sum(...) adds up the areas of all the bars, giving the total area under the histogram.
np.sum(...) 将所有条形的面积相加，得到直方图下的总面积。
● The result is stored in the variable area.
结果存储在变量 area 中。

```python
import numpy as np

# 生成正态分布数据
values = np.random.normal(0, 1, 1000)

# 计算直方图面积（验证归一化）
counts, bins, _ = plt.hist(values, bins=100, density=True)
area = np.sum(counts * np.diff(bins))
print(f"直方图面积 = {area:.4f}")  # 输出应接近1.0
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# ======================
# 1. 生成双峰分布数据
# ======================
np.random.seed(42)  # 固定随机种子保证可复现
group1 = np.random.normal(loc=-2, scale=0.8, size=600)  # 第一个峰（均值-2，标准差0.8）
group2 = np.random.normal(loc=3, scale=1.2, size=400)   # 第二个峰（均值3，标准差1.2）
values = np.concatenate([group1, group2])               # 合并成双峰数据

# ======================
# 2. 计算KDE曲线
# ======================
density = stats.gaussian_kde(values)                    # 核密度估计
x_range = np.linspace(min(values)-2, max(values)+2, 1000)  # 生成评估点范围
y_kde = density(x_range)                                # 计算KDE值

# ======================
# 3. 可视化
# ======================
plt.figure(figsize=(12, 6))

# 绘制直方图（概率密度形式）
plt.hist(values, bins=50, density=True, alpha=0.6,
         color='steelblue', edgecolor='white',
         label='Histogram (normalized)')

# 绘制KDE曲线
plt.plot(x_range, y_kde, color='#FF7F00', linewidth=3,
         label='Kernel Density Estimation')

# 标记原始分布参数
plt.axvline(x=-2, color='red', linestyle='--', alpha=0.5, label='True mean: -2')
plt.axvline(x=3, color='green', linestyle='--', alpha=0.5, label='True mean: 3')

# 图形装饰
plt.title('Bimodal Distribution Analysis\n(Histogram + KDE)', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.3)
plt.xlim(min(x_range), max(x_range))

plt.tight_layout()
plt.show()
```

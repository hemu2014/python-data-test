## Parsing raw HTML strings

Another useful pandas method is `read_html()`. This method will read HTML tables from a given URL, a file-like object, or a raw string containing HTML, and return a list of `DataFrame` objects.

Let's try to read the following `html_string` into a `DataFrame`.

此方法将从给定的 URL、文件对象或包含 HTML 的原始字符串中读取 HTML 表格，**并返回一个 DataFrame 对象的列表。**

```python
html_string = """
<table>
    <thead>
      <tr>
        <th>Order date</th>
        <th>Region</th>
        <th>Item</th>
        <th>Units</th>
        <th>Unit cost</th>
      </tr>
    </thead>
    <tbody>
      <tr>
        <td>1/6/2018</td>
        <td>East</td>
        <td>Pencil</td>
        <td>95</td>
        <td>1.99</td>
      </tr>
      <tr>
        <td>1/23/2018</td>
        <td>Central</td>
        <td>Binder</td>
        <td>50</td>
        <td>19.99</td>
      </tr>
      <tr>
        <td>2/9/2018</td>
        <td>Central</td>
        <td>Pencil</td>
        <td>36</td>
        <td>4.99</td>
      </tr>
      <tr>
        <td>3/15/2018</td>
        <td>West</td>
        <td>Pen</td>
        <td>27</td>
        <td>19.99</td>
      </tr>
    </tbody>
</table>
"""
```

```python
dfs = pd.read_html(html_string)
```

The `read_html` just returned one `DataFrame` object:

```python
len(dfs)
```

```python
print(dfs, type(dfs))
```

```python
df = dfs[0]

df
```

Previous `DataFrame` looks quite similar to the raw HTML table, but now we have a `DataFrame` object, so we can apply any pandas operation we want to it.

```python
df.shape
```

```python
df.loc[df['Region'] == 'Central']
```

```python
df.loc[df['Units'] > 35]
```

### Defining header

Pandas will automatically find the header to use thanks to the <thead> tag.
    
But in many cases we'll find wrong or incomplete tables that make the `read_html` method parse the tables in a wrong way without the proper headers.

To fix them we can use the `header` parameter.

```python
html_string = """
<table>
  <tr>
    <td>Order date</td>
    <td>Region</td>
    <td>Item</td>
    <td>Units</td>
    <td>Unit cost</td>
  </tr>
  <tr>
    <td>1/6/2018</td>
    <td>East</td>
    <td>Pencil</td>
    <td>95</td>
    <td>1.99</td>
  </tr>
  <tr>
    <td>1/23/2018</td>
    <td>Central</td>
    <td>Binder</td>
    <td>50</td>
    <td>19.99</td>
  </tr>
  <tr>
    <td>2/9/2018</td>
    <td>Central</td>
    <td>Pencil</td>
    <td>36</td>
    <td>4.99</td>
  </tr>
  <tr>
    <td>3/15/2018</td>
    <td>West</td>
    <td>Pen</td>
    <td>27</td>
    <td>19.99</td>
  </tr>
</table>
"""
```


代码	返回类型	用途
**pd.read_html(html_string)	List[DataFrame]	获取所有表格的列表**
**pd.read_html(html_string)[0]	DataFrame	获取第一个表格（常用场景）**
关键点：***read_html() 的设计是为了兼容 HTML 中可能存在多个表格的情况，因此始终返回列表。通过 [0] 可以快速提取目标表格。***

```python
pd.read_html(html_string)[0]  # 可能存在多个表格，返回数据是dataframe的列表，访问相应的数据帧需指定下标
```

In this case, we'll need to pass the row number to use as header using the `header` parameter.

```python
pd.read_html(html_string, header=0)[0]
```

###注意header参数， 之前的数据会被丢失

```python
pd.read_html(html_string, header=3)[0]
```

### Complex example

We can also use the `requests` module to get HTML code from an URL to parse it into `DataFrame` objects.

If we look at the given URL we can see multiple tables about The Simpsons TV show.

We want to keep the table with information about each season.

```python
import requests

html_url = "https://en.wikipedia.org/wiki/The_Simpsons"
```

```python
req = requests.get(html_url)
req.text
```

```python
type(req.text)
```

```python
r = requests.get(html_url)

wiki_tables = pd.read_html(r.text, header=0)
```

```python
len(wiki_tables)
```

```python
wiki_tables[1].head()
```

```python
simpsons = wiki_tables[2]
```

```python
simpsons.head()
```

Quick clean on the table: remove extra header rows and set `Season` as index.

```python
simpsons.head()
```

```python
simpsons.set_index('Season', inplace=True)
```

```python
simpsons.head()
```

###使用 .at[] 或 .iat[]（单值快速访问）


场景	推荐方法	优点
按***标签选择单值***	**.at[行标签, '列名']	速度快，代码清晰**
按***位置选择单值***	**.iat[行位置, 列位置]	适合固定位置访问**

```python
# 获取行标签为0、列名为 'Name' 的单个值
most_hits_player.at[0, 'Name']
```

```python
# 获取第0行、第1列的值（假设 'Name' 是第1列）
most_hits_player.iat[0, 1]
```

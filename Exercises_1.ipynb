```python
import matplotlib.pyplot as plt

map = sales.groupby('Year')['Revenue'].sum()
total_revenue = map.sum() # Calculate total revenue

def my_autopct(pct):
    value = int(round(pct*total_revenue/100.0))  # Calculate specific value
    return '{p:.2f}%  (${v:,.0f})'.format(p=pct,v=value)  # Format output string

# Draw pie chart
map.plot(kind='pie', figsize=(6,6), autopct=my_autopct)

# Add title, display total revenue
plt.title('Total Revenue: ${:,.2f}'.format(total_revenue))

plt.show()
```
